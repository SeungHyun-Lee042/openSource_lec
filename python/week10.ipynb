```python
# 10주차 실습 2(최대값 구하기)
# 숫자열을 입력 받아 최대값을 찾아내는 프로그램

num = input('Numbers? ')
cal = []
for i in num.split():
    cal.append(int(i))
result = max(cal)
print('Max = {0:d}'.format(result))
```

```python
# 10주차 실습 3(암호화/복호화)
# 입력 받은 문자열을 암호화하고 복호화하는 프로그램

def enc(str) :
    result = ""
    for i in range(len(str)):
        for j in str.pop():
            asc = ord(j)
            asc += 1
            result = result + chr(asc)
        result += ' '

    return result

def dec(str) :
    result = ""
    for i in range(len(str)):
        for j in str.pop():
            asc = ord(j)
            asc -= 1
            result = result + chr(asc)
        result += ' '

    return result

while(True):
    menu = int(input("원하는 기능을 선택하세요. \n 1) 암호화  2) 복호화  3) 종료  => "))

    if(menu == 1):
        code = input("문자열을 입력하세요 : ")
        print(enc(code.split())+'\n')
    elif(menu == 2):
        code = input("문자열을 입력하세요 : ")
        print(dec(code.split())+'\n')
    elif(menu == 3):
        print('종료합니다. ')
        break
    else:
        print('잘못 입력하셨습니다. \n')
        continue
```

```python
# 10주차 실습 4(전화번호부)
# 10명의 이름과 전화번호를 입력받아 전화번호 사전을 만들고, 이름과 전화번호로 검색하는 프로그램

dic = {}
for i in range(10):
    name, phone = input("이름과 전화번호를 입력해주세요 : ").split(" ")
    dic[name] = phone

while(True):
    menu = int(input("원하는 기능을 선택하세요. \n 1) 이름으로 검색  2) 전화번호로 검색  3) 종료  => "))

    if(menu == 1):
        in_name = input("이름을 입력하세요 : ")
        print("전화번호 : "+dic.get(in_name)+"\n")
    elif(menu == 2):
        in_phone = input("전화번호를 입력하게요 : ")
        temp = [k for k, v in dic.items() if v == in_phone]
        print("이름 : " + temp[0] + "\n")
    elif(menu == 3):
        print('종료합니다. ')
        break
    else:
        print('잘못 입력하셨습니다. \n')
        continue
```
