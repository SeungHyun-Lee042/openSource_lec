```python
# 11주차 실습 7(wc)
# unix의 wc 명령어를 python으로 구현

cp, name = input("입력하세요 : ").split()

file1 = open(name, 'r', encoding='UTF8')
data = file1.readlines()
file1.close()

word = []
for i in data:
    temp = i.split()
    for j in temp:
        word.append(j)

print('라인 수 : '+str(len(data)))
print('단어 수 : '+str(len(word)))
```

```python
# 11주차 실습 8(사칙 연산 클래스)
# 사칙연산 함수 완성하기

class Calc:
    def sum(self, a, b):
        result = a+b
        print("{0:s} + {1:s} = {2:s} 입니다.".format(str(a), str(b), str(result)))

    def subtraction(self, a, b):
        result = a-b
        print("{0:s} - {1:s} = {2:s} 입니다.".format(str(a), str(b), str(result)))

    def multiplication(self, a, b):
        result = a*b
        print("{0:s} * {1:s} = {2:s} 입니다.".format(str(a), str(b), str(result)))

    def division(self, a, b):
        result = a/b
        print("{0:s} / {1:s} = {2:s} 입니다.".format(str(a), str(b), str(result)))


calc = Calc()
calc.sum(1, 2)
calc.subtraction(5,1)
calc.multiplication(2,3)
calc.division(6,2)
```

```python
# 11주차 실습 9(계산기)
# 공백으로 구분되어 있는 계산식을 입력하면 스택을 이용하여 계산을 수행하고 그 결과를 출력하는 프로그램
class stack:
    def __init__(self):
        self.data = []

    def is_Empty(self):
        return True if not self.data else False

    def size(self):
        return len(self.data)

    def push(self, item):
        return self.data.append(item)

    def pop(self):
        return self.data.pop()

    def peek(self):
        return self.data[-1]

operator = {'(': 4, ')': 4, '*': 2, '/': 2, '+': 3, '-': 3}

def cal(t1, t, t2):
    t1 = int(t1)
    t2 = int(t2)
    if t == '+':
        return t1 + t2
    elif t == '-':
        return t1 - t2
    elif t == '*':
        return t1 * t2
    elif t == '/':
        return t1 // t2

calc = input("계산식을 입력하세요. =>  ")
t_input = calc.split()
operater = stack()
result = stack()
num = []

for t in t_input:
    if t != '+' and t != '-' and t != '*' and t != '/':
        result.push(t)
    elif operater.is_Empty():
        operater.push(t)
    elif operator[t] < operator[operater.peek()]:
        operater.push(t)
    else:
        while  not operater.is_Empty() and operator[t] >= operator[operater.peek()]:
            result.push(operater.pop())
        operater.push(t)

while not operater.is_Empty():
    temp = operater.pop()
    result.push(temp)

for i in range(result.size()):
    num.append(result.pop())
num.reverse()

for t in num:
    if t != '+' and t != '-' and t != '*' and t != '/':
          operater.push(t)
    else:
        t2 = operater.pop()
        t1 = operater.pop()
        operater.push(cal(t1, t, t2))
else:
    for i in t_input :
        print(i, end='')
    print(' = {}'.format(operater.pop()))
```
