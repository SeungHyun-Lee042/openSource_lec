```python
# 13주차 실습 1(생일 기온 시각화)
#자신 생일의 서울 지역 최고 기온과 최저 기온을 아래와 같이 시각화하시오.

import matplotlib.pyplot as plt
import csv

f = open('10y_seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

high = []
low = []
for row in data:
    if row[4] != '' and row[3] != '':
        if row[0].split('-')[1] == '04' and row[0].split('-')[2] == '20' :
            high.append(float(row[4]))
            low.append(float(row[3]))

plt.title('04.20 Temperatures')
plt.plot(high, color='red')
plt.plot(low, color='blue')
plt.show()
```

```python
# 13주차 실습 2(최고 기온 시각화)
#서울 지역의 1월과 8월 최고 기온 데이터를 아래와 같이 히스토그램으로 시각화하시오.

import matplotlib.pyplot as plt
import csv

f = open('10y_seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

january = []
august = []
for row in data:
    if row[4] != '':
        if row[0].split('-')[1] == '01':
            january.append(float(row[4]))
        elif row[0].split('-')[1] == '08':
            august.append(float(row[4]))

plt.title('January and August maximum temperature')
plt.hist(august, bins=100, color='red', label='Aug')
plt.hist(january, bins=100, color='blue', label='Jan')
plt.legend()
plt.show()
```
