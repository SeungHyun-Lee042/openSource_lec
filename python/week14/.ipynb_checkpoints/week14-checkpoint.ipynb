```python
# 14  1( )
#''                   .

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import font_manager, rc

cal = []
stand_dev = []
data = pd.read_csv("./population.csv",  encoding='cp949')
route = data[''].tolist()
route.remove('  (1100000000)')
df = data.loc[1:26]
df = df.drop(['202012__'], axis=1)
df = df.drop(['202012__'], axis=1)
df = df.drop([''], axis=1)

for i in df.columns:
    df[i] = df[i].str.replace(",", "")
    df[i] = pd.to_numeric(df[i])
gangjin = df.loc[5]

for i in df.axes[0]:
    temp = df.loc[i]
    loc_cal = gangjin.sub(temp)
    mul = (loc_cal**2)
    stand = np.std(mul)
    stand_dev.append(stand)
    sum = np.sum(mul, axis=0)
    cal.append(sum)
cal.remove(0)
stand_dev.remove(0.0)
route.remove('  (1121500000)')
x = [0, 20, 40, 60, 80, 100]

stand_dif = cal.index(max(cal))
stand_sim = cal.index(min(cal))
sim_str = str(route[stand_sim]) + "\n"+" : " +str(cal[stand_sim]/101)
dif_str = str(route[stand_dif]) + "\n"+ " : " +str(cal[stand_dif]/101)

plt.rc('font', family='NanumGothic')

font_path="C:/Windows/Fonts/HanSantteutDotumRegular.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

plt.figure(figsize=(10,6))
plt.title('   /    ')
plt.plot(np.arange(0, 101, 1), df.loc[stand_sim+2], color='blue', label=sim_str)
plt.plot(np.arange(0, 101, 1), gangjin, color='red', label='  (1121500000)')
plt.plot(np.arange(0, 101, 1), df.loc[stand_dif+2], color='green', label=dif_str)
plt.grid(True)
plt.legend()
plt.show()
```
