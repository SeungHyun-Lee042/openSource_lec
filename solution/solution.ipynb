```python
#기말고사
#문제 1
class Point :
    a =0
    b =0
    def __init__(self, x, y):
        self.a=x
        self.b=y

    def show(self):
        print("("+str(self.a) +"," + str(self.b)+ ")")

    def __add__(self, other):
        self.a = self.a + other.a
        self.b = self.b + other.b
        return self


a = Point(1,2)
a.show()

b = Point(3,4)
b.show()

a = a+b
a.show()

```

```python
import csv

f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

max_date = ""
tem  = []
sum = 0.0
count = 1

for row in data:
    if count % 7 == 0:
        tem.append(sum)
        sum =0
    sum = sum + (float(row[4]) - float(row[3]))
    count = count+1

ind = max(tem)
result = tem.index(ind)
data = result*7 + 1
print("Max Week = 5/"+str(data)+ " ~ 5/" + str(data+6))
```
